## Power spectra of velocity slices

```python
from astropy.io import fits
hdu, = fits.open("/Users/will/Dropbox/SPMFEB13/OrionS/rough-cube.fits")
hdu.header
```

```python
hdu.data.shape
```

```python
nv, ny, nx = hdu.data.shape
pix_scale = 3600*hdu.header['CD1_1']
pix_scale
```

```python
spec = hdu.data.sum(axis=-1).sum(axis=-1)
plot(spec[74:95])
```

```python
thin_map = hdu.data[80,:,:]
thick_map = hdu.data[:,:,:].sum(axis=0)
x = pix_scale*np.arange(nx)
plot(x, thin_map[5], '-o')
plot(x, thick_map[5]/5)
xlim(150.0, 200.0)
gcf().set_size_inches(15, 6)
```

```python
import numpy as np
# units of 1/arcsec
nk = 1 + nx//2
krange = np.arange(nk) / (nx*pix_scale) 
pow_thicks = np.empty((ny, nk))
for i, thick in enumerate(thick_map):
    pow_thick = np.abs(np.fft.rfft(thick))**2
    loglog(krange, krange**3.0 * pow_thick, '-', lw=0.3, alpha=0.2)
    pow_thicks[i,:] = pow_thick
loglog(krange, 4*2.2e7*krange**1.6, 'r')
loglog(krange, 4*2e6*krange**-1.0, 'g')
xlabel('Wavenumber, 1/arcsec')
ylabel('Thick slice power spectrum, $k^3 P(k)$')
xlim(0.01, 1.0)
ylim(1e6, 8e8)
gcf().set_size_inches(10, 6)
```

```python
thin_maps = hdu.data[74:95,:,:]
pow_thins = np.zeros((ny, 1 + nx//2))
for thin_map in thin_maps:
    for irow, thin in enumerate(thin_map):
        pow_thins[irow,:] += np.abs(np.fft.rfft(thin))**2
for pow_thin in pow_thins:
    loglog(krange, krange**3.0 * pow_thin, '-', lw=0.5, alpha=0.4)
loglog(krange, 2.2e7*krange**1.8, 'r')
loglog(krange, 2e6*krange**-1.0, 'g')
xlabel('Wavenumber, 1/arcsec')
ylabel('Thin slice power spectra, $k^3 P(k)$')
xlim(0.01, 1.0)
ylim(1e6, 1e8)
gcf().set_size_inches(10, 6)
```

Thin slices have a shallower slope at the smallest scales: -3 + 1.8 = -1.2

This is only over 1-2.5 arcsec however, or 2 to 5 pixels. 

I think that the red line is just noise.

```python
x1, x2 = 0.01, 1.0
y1, y2 = 1.0e5, 1.0e9
window = [[np.log10(x1), np.log10(x2)], [np.log10(y1), np.log10(y2)]]

f, (ax1, ax2) = subplots(2, 1, sharex=True)

x = krange[None,:]*np.ones_like(pow_thins)
y = pow_thins*x**3
m = (x >= x1) & (x <= x2) & (y >= y1) & (y <= y2)
w = 1./(0.1 + x)
ax1.hist2d(np.log10(x[m]).ravel(), np.log10(y[m]).ravel(), 
       bins=30, weights=w[m].ravel(), range=window)
ax1.set_ylabel('Thin slice power spectra, $k^3 P(k)$')

y = pow_thicks*x**3
m = (x >= x1) & (x <= x2) & (y >= y1) & (y <= y2)
ax2.hist2d(np.log10(x[m]).ravel(), np.log10(y[m]).ravel(), 
       bins=30, weights=w[m].ravel(), range=window)
ax2.set_xlabel('Wavenumber, 1/arcsec')
ax2.set_ylabel('Thick slice power spectra, $k^3 P(k)$')
f.set_size_inches(6,6)
ax1.set_title('[O III] Orion S - VCA for Thin vs Thick slices')
f.tight_layout()
f.savefig("will-vca-test.pdf")
```
